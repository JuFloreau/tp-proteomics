```python
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import pandas as pd
```

```python
fig, ax = plt.subplots()
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)

ax.plot(x, norm.pdf(x),
       'r-', lw=5, alpha=0.6)

ax.plot(x, np.full(len(x), 0.2),
       'b-', lw=1)

fig.show()
```

```python
file=pd.read_csv("data/TCL_wt1.tsv",sep="\t")
```

```python
file
```

```python
df=file.dropna()
df
```

```python
df.iloc[:,3:7]=df.iloc[:,3:7].astype('float')
df.dtypes
```

```python
log2=df['Log2 Corrected Abundance Ratio']
log2.describe()
```

```python
mean=log2.mean()
var=log2.shape[0]/(log2.shape[0]-1)*np.var(log2)
mean,var
```

```python
from math import sqrt
fig, ax = plt.subplots()
hist = ax.hist(log2, bins=100) # draw histogram
x = np.linspace(min(log2), max(log2), 100) # generate PDF domain points
dx = hist[1][1] - hist[1][0] # Get single value bar height
scale = len(log2)*dx # scale accordingly
ax.plot(x, norm.pdf(x, mean, sqrt(var))*scale) # compute theoritical PDF and draw it
fig.show()
fig.savefig("picture/histogram_log2FC.png")
```

```python
log10pvalue=df['-LOG10 Adj.P-val']
log10pvalue
```

```python
fig, ax = plt.subplots()
ax.scatter(x=log2,y=log10pvalue)
ax.vlines(mean+sqrt(var),0,8)
ax.hlines(-np.log10(0.001),-4,1)
fig.show()
fig.savefig("picture/volcanoplot.png")
```

```python
df_ORA=df.loc[(df['Log2 Corrected Abundance Ratio']>mean+sqrt(var)) & (df['-LOG10 Adj.P-val']>-np.log10(0.001))]
```

```python
from xml.etree.ElementTree import parse, dump
# Parse the E.Coli proteome XML Document
tree = parse('data/uniprot-proteome_UP000000625.xml')
root = tree.getroot()
ns = '{http://uniprot.org/uniprot}' # MANDATORY PREFIX FOR ANY SEARCH within document
# Store all entries aka proteins in a list of xml nodes
proteins = root.findall(ns + 'entry')
# Display the xml subtree of the first protein 
dump(proteins[0])
```

```python
# Find the xml subtree of a protein with accession "P31224"
for acc_num in df_ORA["Accession"]:
    for entry in proteins:
        accessions = entry.findall(ns+"accession")
        for acc in accessions:
            if acc.text == acc_num:
                dump(entry)
                break
```
